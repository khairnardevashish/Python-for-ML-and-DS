## Example

```python
import numpy as np 
import pandas as pd

# Defining the three dataframes indicating the gold, silver, and bronze medal counts
# of different countries
gold = pd.DataFrame({'Country': ['USA', 'France', 'Russia'],
                         'Medals': [15, 13, 9]}
                    )
silver = pd.DataFrame({'Country': ['USA', 'Germany', 'Russia'],
                        'Medals': [29, 20, 16]}
                    )
bronze = pd.DataFrame({'Country': ['France', 'USA', 'UK'],
                        'Medals': [40, 28, 27]}
                    )
```

```python

df = pd.merge(silver, gold, how ="outer", on = "Country")
df_1 = pd.merge(df, bronze, how ="outer", on = "Country")
df_1 = df_1.fillna(0)
df_1["Medals"] = df_1["Medals_x"] + df_1["Medals_y"] + df_1["Medals"]
df_1.sort_values(by=["Medals"], ascending= False)
df_1.set_index('Country', inplace = True)
df_1 = df_1.loc[:, ['Medals']]
df_1 = df_1.sort_values(by=["Medals"], ascending= False)
print(df_1)
```

## Lambda function & Pivot tables

```python
df.head()
```

```python
def is_rainy(x):
    return x>0

df["is_rainy"] = df["rain"].apply(is_rainy)
```

```python
df.head()
```

```python
df["strong_wind"] = df["wind"].apply(lambda x : x>2)
```

```python
df.head()
```

```python

df
```

```python
group_rain = df.groupby('rain')
group_rain.is_rainy.mean()
```
